```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('country_wise_accuracy.csv')
dataframe.head()
```

```python
dataframe.keys()
```

```python
performance = pd.DataFrame({'Model': ['LogisticRegression',
                                                     'RandomForest',
                                                     'KNN',
                                                     'DecisionTree',
                                                     'LinearSupportVector',
                                                     'XGBOOST',
                                                     'GaussianNaiveBayes',
                                                     'AdaBoost'],
                                           'Accuracy': [dataframe["LogisticRegression"].mean(),
                                                        dataframe["RandomForest"].mean(),
                                                        dataframe["KNN"].mean(),
                                                        dataframe["DecisionTree"].mean(),
                                                        dataframe["LinearSupportVector"].mean(),
                                                        dataframe["XGBOOST"].mean(),
                                                        dataframe["GaussianNaiveBayes"].mean(),
                                                        dataframe["AdaBoost"].mean()]})
```

```python
performance.sort_values(by = "Accuracy", ascending = False)
```

```python
performance = performance.sort_values(by = "Accuracy", ascending = False)
```

```python
model_name = performance['Model'].tolist()
model_accuracy = performance['Accuracy'].tolist()

x_pos = np.arange(len(model_name))
plt.bar(x_pos, model_accuracy)

# Add title and axis names
plt.title('Average Accuracy Plot across All Countries')
plt.xlabel('Models')
plt.ylabel('Accuracy')

# Create names on the x axis
plt.xticks(x_pos, model_name)
plt.xticks(rotation = 65)

# Show graph
plt.show()
```

```python
# Horizontal Plot
performance = performance.sort_values(by = "Accuracy", ascending = True)
model_name = performance['Model'].tolist()
model_accuracy = performance['Accuracy'].tolist()

plt.barh(model_name, model_accuracy)
plt.title('Average Accuracy Plot across All Countries')
plt.ylabel('Model')
plt.xlabel('Accuracy')
plt.show()
```
